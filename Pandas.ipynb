# 중복데이터 확인 및 처리 

```python
data = {'key1':['a', 'b', 'b', 'c', 'c'], 'key2':['v', 'w', 'w', 'x', 'y'], 'col':[1, 2, 3, 4, 5]}
df = pd.DataFrame(data, columns=['key1','key2','col'])

df
```

## 중복데이터 확인하기 
- DataFrame.duplicated()

```python
df.duplicated(['key1']) # key1을 기준으로 1,2와 3,4가 중복임 (이때 처음의 값은 false로 나온다.)
```

```python
df.duplicated(['key1','key2']) # key1, key2를 동시에 기준은로 하면 index 1,2가 서로 중복 (두가지 기준에 대해 모두 중복인 것)
```

- keep = fist, last, False

```python
df.duplicated(['key1'], keep='first') # default. 처음값의 duplicate 여부를 false로 나머지는 true 반환
```

```python
df.duplicated(['key1'], keep='last') # first의 반대. 마지막의 duplicate 여부를 true, 나머지 false 반환
```

```python
df.duplicated(['key1'], keep=False) # 처음, 끝값 여부 고려하지 않고 중복이면 무조건 True 반환 
# => 나중에 drop_duplicates()에서 keep할 생각이 없다는 뜻
```

## 중복데이터 처리하기
- drop_duplicates()

```python
df.drop_duplicates(['key1'], keep='first') # keep의 파라미터 값에 따라서 unique한 값 한개만 남기고 중복을 제거
```

```python
df.drop_duplicates(['key1'], keep='last')
```

```python
df.drop_duplicates(['key1'], keep=False)
```

```python
# else if가 많이 나와야하는 경우나 swith case의 경우에는 위와 같은 코드가 복잡해지므로 이에 대해서는 하나의 함수 만들어 사용

# 나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else: cat= 'Elderly'

    return cat

# lambda식에 위에 생성한 get_category() 함수를 반환값으로 지정
# get_categoty(X)는 입력값으로 'Age'칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head(5)
        
```

```python
# 탈퇴회원수에 대한 피봇테이블
pivot = pd.pivot_table(df, values = '탈퇴회원수' , index = ['가입월'], columns = ['탈퇴월'])
pivot
```

```python
import random
a = []
b = []

for i in range(100):
    a.append(random.randint(1,3))
    b.append(random.randint(1,3))
df = pd.DataFrame({'품목' : a, '크기' : b})
```

```python
df['금액'] = df['품목'] * df['크기'] * 500
df['수수료'] = df['금액'] * 0.1
df
```

```python
fruit_name = {1 : '토마토', 2 : '바나나', 3 : '사과'}
fruit_size = {1 : '소', 2 : '중', 3 : '대'}

df['품목'] = df['품목'].map(fruit_name)
df['크기'] = df['크기'].map(fruit_size)
```

```python
# 항목 별, 크기 별로 판매 개수와 판매 금액의 합 구하기

pd.pivot_table(df, values = '금액', index = ['품목'], columns = ['크기'], aggfunc = ( 'count', 'sum'))
```

```python
# 항목 별, 크기 별로. 판매 개수와 판매 금액 / 수수료의 합

pd.pivot_table(df, index = ['품목'], columns = ['크기'], aggfunc = {'금액' : ['count', 'sum'], '수수료' : 'sum'})
```
